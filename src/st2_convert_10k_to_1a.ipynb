Create a code that moves all files within a folder outside subfolders to another subfolder called Q1.

```python
import os
import shutil

yearrange = list(range(2006, 2023, 1))
for year in yearrange:
    # Source folder path
    source_folder = f"/Users/pedrovallocci/Documents/PhD (local)/Research/By Topic/Measuring knowledge capital risk/output/1A files/{year}/"

    # Destination folder path
    destination_folder = f"{source_folder}/Q1"

    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate over the files in the source folder
    for root, _, files in os.walk(source_folder):
        for file_name in files:
            # Get the full path of the file
            file_path = os.path.join(root, file_name)

            # Check if the file is not in a subfolder
            if root == source_folder:
                # Move the file to the destination folder
                shutil.move(file_path, destination_folder)
                print(f"Moved {file_name} to {destination_folder}")

print("All files moved successfully!")
```
